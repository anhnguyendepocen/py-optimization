### Graphing incoming data

```python
# make sure you include this at the top of any code cell that runs the graphs:
%matplotlib notebook  

# in order to actually capture your data and be able to save it, you
# must create a variable to store your CollectOpticsData function:
run1 = CollectOpticsData()
```

```python
# now, to make the magic happen, run your variable as a function:
# note that you can select which detectors to graph

run1(detectors=['A', 'B', 'B_', 'C1'])    # detectors A, B, B', and Coincidence1
```

```python
# once youve finished collecting, to access your data you 
# must call the `data` property of your variable

run1.data
```
